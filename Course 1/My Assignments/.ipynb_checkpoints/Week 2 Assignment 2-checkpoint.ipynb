```python
import math
import numpy
```

```python
# Рассмотрим сложную математическую функцию на отрезке [1, 15]:
def f(x):
    return math.sin((x/5.))*math.exp((x/10.)) + 5.*math.exp((-1.*x/2.))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
res = numpy.arange((15), dtype = float).reshape(15,1)
for i in range(0,15):
    res[i] = f(i+1)
```

```python
x=numpy.arange(0.,15.,1.) #Массив значений аргумента
plt.plot(x, res) #Построение графика
plt.show() #Показать график
```

```python
#Сформируйте систему линейных уравнений
#(то есть задайте матрицу коэффициентов A и свободный вектор b)
#для многочлена первой степени, который должен совпадать с функцией f в точках 1 и 15
```

```python
A  = numpy.arange((4), dtype = float).reshape(2,2)
b = numpy.arange((2), dtype = float).reshape(2,1)
```

```python
A = [[1.,1.],[1., 15.]]
b[0][0] = res[0]
b[1][0] = res[14]
det = numpy.linalg.det(A)
if (det == 0):
    print('Система не имеет решений')
else:
    r = numpy.linalg.solve(A,b)
    print(" ".join(str(x) for x in r))
```

```python
def func(x, c):
    return c[0]+c[1]*x

x=numpy.arange(0.,15.,1.) #Массив значений аргумента
plt.plot(x, func(x, r)) #Построение графика
plt.plot(x, res) #Построение графика

plt.show() #Показать график
```

```python
#Повторите те же шаги для многочлена второй степени, который совпадает с функцией f в точках 1, 8 и 15. 
```

```python
A  = numpy.arange((9), dtype = float).reshape(3,3)
b = numpy.arange((3), dtype = float).reshape(3,1)
```

```python
A = [[1.,1., 1.**2],[1., 8., 8.**2],[1., 15., 15.**2]]
b[0][0] = res[0]
b[1][0] = res[7]
b[2][0] = res[14]
```

```python
det = numpy.linalg.det(A)
if (det == 0):
    print('Система не имеет решений')
else:
    r = numpy.linalg.solve(A,b)
    print(" ".join(str(x) for x in r))
```

```python
def func(x, c):
    return c[0]+c[1]*x+c[2]*x**2

x=numpy.arange(0.,15.,1.) #Массив значений аргумента
plt.plot(x, func(x, r)) #Построение графика
plt.plot(x, res) #Построение графика

plt.show() #Показать график
```
