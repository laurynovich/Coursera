```python
import pandas as pd
import numpy as np
import math as m
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import gamma
%matplotlib inline
```

Выберите ваше любимое непрерывное распределение. 

$\textbf{Было выбрано Гамма распределение $\Gamma$($\kappa$,$\theta$)
c параметрами $\kappa$ = 9 , $\theta$ = 0.5}$

```python
k = 9
teta = 0.5
```

Сгенерируйте из него выборку объёма 1000, постройте гистограмму выборки и нарисуйте поверх неё теоретическую плотность распределения вашей случайной величины.

```python
x = np.linspace(0,10,100)
func = sts.gamma(k,0,teta)
sample = func.rvs(1000)
```

Гистограмма выборки:

```python
plt.hist(sample, density = True, label='sample histogram')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
pdf = func.pdf(x)
plt.plot(x, pdf, label='theoretical PDF')
plt.legend(loc='upper right')
```

$\textbf{Теоретические значения для Гамма распределения: }$

среднее = $\kappa$$*$$\theta$, дисперсия = $\kappa$$*$$\theta$$^2$

```python
mu = k * teta
d = k * teta * teta
```

Сгенерируйте 1000 выборок объёма n и постройте гистограммы распределений их выборочных средних

# n = 5 

```python
n = 5
```

```python
#Генерация выборки и вычисление среднего для выборки
med_5 = np.arange(1000, dtype = 'float').reshape(1000,)
for i in range(0, len(med_5)):
    sample_i = func.rvs(n)
    med_5[i] = np.mean(sample_i)
```

Построение графиков для n = 5

```python
#Построение гистограммы средневыборочных
plt.hist(med_5, density = True, label = 'Sample mean histogram')
plt.ylabel('fraction of samples')
plt.xlabel('$x$')

# Вычисление pdf нормального распределения с данными параметрами (каждый раз одинково)
norm_rv = sts.norm(mu, m.sqrt(d/n))
pdf = norm_rv.pdf(x)
#Построение на том же графике
plt.plot(x, pdf, label='Norm theoretical pdf')

# Легенда графика
plt.legend(loc='upper right')
```
